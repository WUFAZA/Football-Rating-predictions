**PREDICTING FIFA RANKINGS**

# **Importing data libraries**

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

# Loading data sets

```python
df = pd.read_csv('https://raw.githubusercontent.com/falaprodavi/Football-Player-Dataset2022/refs/heads/main/datasets/CLEAN_FIFA23_official_data.csv')
df
```

## **Seperating datasets**

```python
y = df['Potential']
y
```

```python
columns_to_add = ['Overall','Contract Valid Until','Height(cm.)','Weight(lbs.)','Release Clause(£)']
x = df[columns_to_add]
x
```

## **Data Splitting**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state= 100)

x_train
```

```python
x_test
```

```python
y_train
```

# **Model Building**

## **Linear Regression**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

## **Testing the Model**

```python
y_old_pred = lr.predict(x_train)
y_new_pred = lr.predict(x_test)
```

## Trying to get the same y values as last time

```python
y_old_pred
```

```python
old_y = y_old_pred.reshape(-1,1)
old_y
```

## Checking for the split up x values

```python
y_new_pred
```

```python
new_y = y_new_pred.reshape(-1,1)
new_y
```

# **Evaluating the model**

```python
from sklearn.metrics import mean_squared_error,r2_score

lr_train_mse = mean_squared_error(y_train, y_old_pred)
r2_score_train = r2_score(y_train, y_old_pred)

lr_test_mse = mean_squared_error(y_test,y_new_pred)
r2_score_test = r2_score(y_test,y_new_pred)

print("LR train mean squared error: ",lr_train_mse)
print("LR r2 train: ",r2_score_train)
print("LR train mean squared error: ",lr_test_mse)
print("LR r2 train: ",r2_score_test)
```

```python
lr_results = pd.DataFrame(['LinearRegression',lr_train_mse,r2_score_train,lr_test_mse,r2_score_test]).transpose()
lr_results.columns = ['Methods','Training Mse','Training R2','Test mse','Test r2']
print(lr_results)
```
